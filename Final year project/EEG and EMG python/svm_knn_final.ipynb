```python
import pandas as pd
import numpy as np
from scipy.stats import skew, kurtosis

# Define custom functions for extracting features
def mav(data): #mean absolute value
    return np.mean(np.abs(data), axis=0)

def wavelength(data): #waveform length
    return np.sum(np.abs(np.diff(data)), axis=0)

def zero_crossing_rate(data):
    return np.sum(np.diff(np.sign(data), axis=0) != 0, axis=0) / (len(data)-1)

def skewness(data):
    return skew(data, axis=0)

def kurt(data):
    return kurtosis(data, axis=0)

def rms(data): #root mean square
    return np.sqrt(np.mean(data**2,axis=0))  

def SSI(data): #Simple Square Integral
    return np.sum(data**2,axis=0)

def abs_diffs_signal(data):
    return np.sum(np.abs(np.diff(data,axis=0)),axis=0)
```

```python
fe_data = subjects_data.agg([ 'min','max', np.ptp, SSI, rms, abs_diffs_signal])

fe_data = fe_data.reset_index()
fe_data.head()
```

```python
#fe1_data = subjects_data.agg([ 'min','max', np.ptp, SSI, rms, abs_diffs_signal, 
#                              mav, wavelength, zero_crossing_rate, skewness, kurt])
fe1_data = subjects_data.agg([ 'min','max', np.ptp, SSI, rms, abs_diffs_signal, 
                              mav, wavelength, skewness, kurt])

fe1_data = fe1_data.reset_index()
fe1_data
```

```python
from sklearn.model_selection import train_test_split
# from sklearn.preprocessing import StandardScaler

# let x as features and y as label (class of hand gesture)
# drop 'label' and 'class' in features because we only need the feature extraction from channel 1--8
x = fe1_data.drop(columns=['label','class'], level=0)
y = fe1_data['class']

# Split the data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=42)

## Data normalization
mean = x_train.mean(axis=0)
std = x_train.std(axis=0)
x_train -= mean
x_train /= std
x_test -= mean
x_test /= std
```

```python
y_train.value_counts()
```

```python
y_test.value_counts()
```

```python
from sklearn.svm import SVC

# Build SVM model
svm_model = SVC(kernel='linear')  # You can choose the kernel based on your data characteristics

# Train the model
svm_model.fit(x_train, y_train)

# Predict on the test set
y_pred_svm = svm_model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report
from sklearn.metrics import precision_score, f1_score, recall_score
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Evaluate svm model
svm_accuracy = accuracy_score(y_test, y_pred_svm)
svm_precision = precision_score(y_test, y_pred_svm, average='weighted')
svm_recall = recall_score(y_test, y_pred_svm, average='weighted')
svm_f1score = f1_score(y_test, y_pred_svm, average='weighted')
svm_confusion_matrix = confusion_matrix(y_test, y_pred_svm)

# Print the evaluation metrics
print(f'Accuracy: {svm_accuracy:.4f}')
print(f'Precision: {svm_precision:.4f}')
print(f'Recall: {svm_recall:.4f}')
print(f'F1 score: {svm_f1score:.4f}')
print('----------------------------')

# Plot the confusion matrix as a heatmap
plt.figure(figsize=(4, 3))
sns.heatmap(svm_confusion_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=svm_model.classes_, yticklabels=svm_model.classes_)
plt.title('SVM - Linear')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```
